```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_nifty = pd.read_excel('/content/sample_data/Assignment 1.1-Interpretation of the data.xlsx', sheet_name='NIFTY')
df_sensex = pd.read_excel('/content/sample_data/Assignment 1.1-Interpretation of the data.xlsx', sheet_name='SENSEX')
```

```python
plt.figure(figsize=(50,25))
plt.bar(df_nifty['Ticker'],df_nifty['CAGR'])
```

**Top 5 in CAGR growth**


```python
top_5_cagr = df_nifty.sort_values('CAGR', ascending=False)
top_5_cagr.head()
```

**Bottom 5 in CAGR growth**


```python
bottom_5_cagr = df_nifty.sort_values('CAGR', ascending=True)
bottom_5_cagr.head()
```

**Highest Returns in 2018**


```python
highest_returns_2018 = df_nifty.groupby('Ticker')['18'].max()
highest_returns_2018.max()
```

```python
highest_18 = df_nifty.sort_values('18', ascending=False)
highest_18
```

```python
plt.figure(figsize=(50,25))
plt.bar(df_nifty['Ticker'],df_nifty['18'])
```

**Highest Returns in 2019**

```python
highest_returns_2019 = df_nifty.groupby('Ticker')['19'].max()
highest_returns_2019.max()
```

```python
highest_19 = df_nifty.sort_values('19', ascending=False)
highest_19
```

```python
plt.figure(figsize=(55,25))
plt.bar(df_nifty['Ticker'],df_nifty['19'])
```

**Highest Returns in 2020**

```python
highest_returns_2020 = df_nifty.groupby('Ticker')['20'].max()
highest_returns_2020.max()
```

```python
highest_20 = df_nifty.sort_values('20', ascending=False)
highest_20
```

```python
plt.figure(figsize=(55,25))
plt.bar(df_nifty['Ticker'],df_nifty['20'])
```

**Average Returns in 2018**

```python
avg_returns_2018 = df_nifty.groupby('Ticker')['18'].mean()
avg_returns_2018.mean()
```

**Average Returns in 2019**

```python
avg_returns_2019 = df_nifty.groupby('Ticker')['19'].mean()
avg_returns_2019.mean()
```

**Average Returns in 2020**

```python
avg_returns_2020 = df_nifty.groupby('Ticker')['20'].mean()
avg_returns_2020.mean()
```
